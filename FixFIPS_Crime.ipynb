```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/JustinHolmes/Desktop/CrimeData.tsv", sep = '\t', header = 0)
dfPOP = pd.read_csv("/Users/JustinHolmes/Desktop/2014PopEstimate.csv", sep = ',', header = 0)
```

```python
df.rename(columns={'P1VLNT':'Violent Crimes', "P1PRPTY":"Property Crimes"}, inplace=True)
```

```python
df2 = df[["FIPS_ST", "FIPS_CTY", "Violent Crimes", "Property Crimes"]].copy()
df2["FIPS"] = 0
df2 = df2[["FIPS","FIPS_ST", "FIPS_CTY", "Violent Crimes", "Property Crimes"]].copy()
df2.head()
```

```python
for index, row in df2.iterrows():
    df2["FIPS"][index] = 1000*row[1]+row[2]
df2.tail()
```

```python
for index, row in dfPOP.iterrows():
    dfPOP["FIPS"][index] = 1000*row[1]+row[2]
```

```python
dfPOP.head()
```

```python
dfCrime = df2[["FIPS", "Violent Crimes", "Property Crimes"]]
dfCrime.head()
```

```python
dfFinal = dfCrime.merge(dfPOP, on = 'FIPS', how = 'left')
dfFinal = dfFinal[["FIPS", "Violent Crimes", "Property Crimes", "2014 Population"]]
dfFinal.head()
```

```python
dfFinal["Violent Crime Rate"] = 0
dfFinal["Violent Crime Rate"] = 0
dfFinal["Violent Crime Rate"] = round(dfFinal["Violent Crimes"]/(dfFinal["2014 Population"]/100000),0)
dfFinal["Property Crime Rate"] = round(dfFinal["Property Crimes"]/(dfFinal["2014 Population"]/100000),0)
dfFinal.head()
```

```python
dfFinal.to_csv("/Users/JustinHolmes/Desktop/2014CrimeData.csv", index = None)
```
